Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ans - Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two numerical variables. They help in identifying patterns, clusters, or trends within the data.
Bar Plot:

Use: Bar plots are used to display and compare the values of categorical variables. They are effective in showing the distribution or count of categories within a dataset.
Histogram:

Use: Histograms are useful for displaying the distribution of a single numerical variable. They represent the frequency or count of data points within predefined intervals (bins) to show the underlying frequency distribution.
Box Plot (Box-and-Whisker Plot):

Use: Box plots are helpful in visualizing the distribution of numerical data through their quartiles. They provide information about the median, spread, and potential outliers in the dataset.
Heatmap:

Use: Heatmaps are effective for visualizing relationships and correlations within a dataset. They use color-coding to represent the magnitude of values in a matrix, making it easy to identify patterns or clusters in large datasets, especially in correlation matrices.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the URL
url = "https://github.com/mwaskom/seaborn-data/raw/master/fmri.csv"
fmri_data = pd.read_csv(url)


# Create a line plot using Seaborn
sns.set(style="darkgrid")  # Set the style for the plot

# Plotting line plot for different events and regions
plt.figure(figsize=(10, 6))  # Set the figure size

# Use Seaborn's lineplot function to create the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Set the title and labels
plt.title('Signal vs Timepoint for Different Events and Regions (fmri dataset)')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the plot
plt.legend(loc='best')  # Show legend
plt.tight_layout()
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from the URL
url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
titanic_data = pd.read_csv(url)
 

# Plot two box plots using seaborn
plt.figure(figsize=(12, 6))

# Box plot for 'age' based on 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare' based on 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Show the plots
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset from the URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
diamonds_data = pd.read_csv(url)


# Plot a histogram for the 'price' column with 'cut' as a hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30, kde=True)
plt.title('Histogram of Price with Cut as Hue')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from the URL
url = "https://github.com/mwaskom/seaborn-data/raw/master/iris.csv"
iris_data = pd.read_csv(url)

# Plot a pair plot with 'species' as hue
sns.pairplot(iris_data, hue='species')
plt.title('Pair Plot of Iris Dataset with Species as Hue')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset from seaborn
url = "https://github.com/mwaskom/seaborn-data/raw/master/flights.csv"
flights_data = pd.read_csv(url)


# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu', cbar_kws={'label': 'Passengers'})
plt.title('Heatmap of Flights Dataset')
plt.show()
```
