Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

# Given data
mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 100  # You need to replace this with the actual sample size

# Calculate standard error of the mean (SEM)
sem = std_dev / (sample_size**0.5)

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * sem

# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

# Interpretation of results
print(f"95% Confidence Interval: {confidence_interval}")
print(f"Interpretation: We are 95% confident that the true population mean lies between {confidence_interval[0]:.2f} and {confidence_interval[1]:.2f}.")
```

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

```python
pip install scipy
```

```python
import numpy as np
from scipy.stats import chisquare

# Replace these values with the actual observed frequencies of each color in your M&Ms bag
observed_blue = 30
observed_orange = 25
observed_green = 20
observed_yellow = 15
observed_red = 10
observed_brown = 50

observed_frequencies = np.array([observed_blue, observed_orange, observed_green, observed_yellow, observed_red, observed_brown], dtype=int)

# Replace this value with the total number of M&Ms in the bag
total_number_of_MMs_in_bag = sum(observed_frequencies)

# Expected frequencies based on the given distribution
expected_frequencies = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20]) * total_number_of_MMs_in_bag

# Perform the chi-square goodness of fit test
chi2_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Print the results
print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_value)

# Compare p-value to the significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The distribution of colors is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution of colors is not significantly different from the expected distribution.")
```

Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following
data:

Group A group b

Outcome 1 - 20 15
Outcome 2 - 10 25
Outcome 3 - 15 20

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create the contingency table
data = np.array([[20, 15], [10, 25], [15, 20]])

# Perform the chi-square test for independence
chi2_stat, p_value, _, _ = chi2_contingency(data)

# Print the results
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Compare the p-value to the significance level (e.g., 0.05) to make a decision
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant association between the groups and outcomes.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the groups and outcomes.")
```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

```python
pip install statsmodels
```

```python
import statsmodels.stats.proportion as prop

# Given data
total_population = 500
smokers = 60

# Calculate the proportion of smokers in the sample
sample_proportion = smokers / total_population

# Set the confidence level (e.g., 95%)
confidence_level = 0.95

# Calculate the confidence interval
confidence_interval = prop.proportion_confint(smokers, total_population, alpha=1-confidence_level, method='normal')

# Print the results
print(f"Sample Proportion: {sample_proportion:.4f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")
```

Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results.

```python
import scipy.stats as stats

# Given data
sample_mean = 75
sample_std_dev = 12
sample_size = 757

# Set the confidence level (e.g., 90%)
confidence_level = 0.90

# Calculate the standard error of the mean
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * standard_error

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the results
print(f"Sample Mean: {sample_mean}")
print(f"90% Confidence Interval: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")
```

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

```python
pip install numpy matplotlib scipy
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Set the degrees of freedom
degrees_of_freedom = 10

# Create a range of x values
x_values = np.linspace(0, 30, 1000)

# Calculate the chi-square probability density function (PDF) for each x value
pdf_values = chi2.pdf(x_values, degrees_of_freedom)

# Plot the chi-square distribution
plt.plot(x_values, pdf_values, label=f'Chi-square Distribution (df={degrees_of_freedom})')

# Shade the area corresponding to a chi-square statistic of 15
x_shaded = np.linspace(0, 15, 100)
plt.fill_between(x_shaded, chi2.pdf(x_shaded, degrees_of_freedom), color='orange', alpha=0.3, label='Chi-square Statistic = 15')

# Add labels and legend
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density Function')
plt.title('Chi-square Distribution')
plt.legend()

# Show the plot
plt.show()
```

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.

```python
pip install statsmodels
```

```python
import numpy as np
from scipy.stats import norm

# Given data
sample_size = 1000
prefer_coke = 520

# Calculate the sample proportion
sample_proportion = prefer_coke / sample_size

# Set the confidence level (e.g., 99%)
confidence_level = 0.99

# Calculate the standard error of the proportion
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the margin of error
margin_of_error = norm.ppf((1 + confidence_level) / 2) * standard_error

# Calculate the confidence interval
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)

# Print the results
print(f"Sample Proportion: {sample_proportion:.4f}")
print(f"{confidence_level * 100}% Confidence Interval: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")
```
