Q1. What is the relationship between polynomial functions and kernel functions in machine learning
algorithms?

ans - Polynomial functions and kernel functions are both used in machine learning algorithms, but they serve different purposes.

Polynomial functions are a class of functions that involve powers and coefficients of a variable. In the context of machine learning, polynomial functions are often used as feature transformations. By applying polynomial functions to the original input features, we can create new features that capture nonlinear relationships between the variables. Polynomial features are useful when the underlying relationship between the variables is not linear and cannot be accurately represented by a linear model.

Kernel functions, on the other hand, are mathematical functions that operate on pairs of data points. They are commonly used in kernel methods, such as Support Vector Machines (SVMs). Kernel functions enable SVMs to implicitly map the input data into a higher-dimensional feature space, where the data becomes linearly separable. This mapping is often nonlinear and can capture complex relationships between the data points. The use of kernel functions allows SVMs to efficiently compute the decision boundary in the transformed feature space without explicitly calculating the transformed features.

The relationship between polynomial functions and kernel functions in machine learning algorithms lies in their ability to capture nonlinear relationships. Both approaches enable models to learn complex patterns and make nonlinear decisions. Polynomial functions achieve this by explicitly transforming the input features into a higher-dimensional space, while kernel functions achieve it by implicitly mapping the data into a higher-dimensional space. However, it's important to note that kernel functions offer more flexibility in capturing complex relationships, as they can leverage various types of kernel functions (e.g., Gaussian, polynomial, sigmoid) to achieve the desired transformation.

Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?

ans - In Support Vector Regression (SVR), epsilon is a hyperparameter that controls the width of the epsilon-insensitive tube around the training samples. The epsilon-insensitive tube is used to define the margin within which errors are considered acceptable and do not contribute to the loss function.

When you increase the value of epsilon in SVR, it typically leads to an increase in the number of support vectors. Support vectors are the data points that lie on the margin or violate the margin boundary. They play a crucial role in defining the decision function and the regression line in SVR.

A larger value of epsilon allows for more errors to be within the acceptable margin, meaning more data points can be considered as support vectors. This occurs because a wider epsilon-insensitive tube permits more training samples to fall within the margin without penalizing them.

Consequently, the model becomes more tolerant to errors, leading to a larger number of support vectors. These additional support vectors help capture the variations and complexities of the data, allowing the regression line to adapt to a wider range of target values.

However, it's important to note that increasing epsilon excessively can lead to overfitting, where the model becomes too flexible and may not generalize well to unseen data. Therefore, choosing an appropriate value for epsilon is crucial and often requires experimentation and validation using techniques like cross-validation.

Q4. How does the choice of kernel function, C parameter, epsilon parameter, and gamma parameter
affect the performance of Support Vector Regression (SVR)? Can you explain how each parameter works
and provide examples of when you might want to increase or decrease its value?

ans - In Support Vector Regression (SVR), the choice of kernel function, C parameter, epsilon parameter, and gamma parameter can significantly affect the performance of the model. Let's discuss each parameter and its impact:

Kernel Function:

The kernel function determines the type of non-linear mapping used to transform the input features.
Different kernel functions, such as linear, polynomial, Gaussian (RBF), and sigmoid, have different properties and capture different types of relationships in the data.
You should choose the kernel function based on the characteristics of your data and the type of relationships you want the model to capture.
For example, a linear kernel is appropriate when the relationship between the features and the target variable is expected to be linear, while a Gaussian kernel is suitable for capturing complex and nonlinear relationships.
C Parameter:

The C parameter, also known as the regularization parameter, controls the trade-off between model simplicity and training error.
A smaller C value implies a softer margin, allowing more training errors but leading to a simpler model.
On the other hand, a larger C value enforces a harder margin, penalizing training errors and potentially resulting in a more complex model.
You might want to increase the C value when you believe overfitting is occurring, as it encourages the model to focus on minimizing training errors.
Conversely, decreasing the C value can be useful when the model is too complex and not generalizing well to new data.
Epsilon Parameter:

The epsilon parameter defines the width of the epsilon-insensitive tube around the training samples in SVR.
It determines the margin within which errors are considered acceptable and do not contribute to the loss function.
A larger epsilon value allows for a wider margin, increasing the tolerance for errors and potentially increasing the number of support vectors.
Increasing epsilon can be useful when you want the model to be more robust to noise or when you have a larger acceptable error threshold.
Conversely, decreasing epsilon tightens the margin and makes the model less tolerant to errors, potentially reducing the number of support vectors.
Gamma Parameter:

The gamma parameter defines the influence of a single training sample on the SVR model.
It controls the shape of the Gaussian (RBF) kernel and the reach of each training sample.
A smaller gamma value makes the Gaussian kernel wider and results in a smoother decision boundary, allowing for a more global fit.
Increasing the gamma value makes the Gaussian kernel narrower and leads to a more localized fit, potentially overfitting the training data.
You might want to increase the gamma value when you suspect that the relationship between the features and the target variable is highly localized or when you want the model to focus on individual data points.
It's important to note that the optimal values for these parameters depend on the specific dataset and problem at hand. Experimentation and validation techniques like cross-validation can help identify the best combination of parameter values that yield good performance and generalization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset.keys()
```

```python
X = pd.DataFrame(dataset.data,columns=dataset.feature_names)
y = dataset.target
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.29, random_state=92)
```

```python
X_train.shape,y_train.shape
```

```python
y_test.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import  tree
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled = True)
```

```python
#post pruning
treeclassifier = DecisionTreeClassifier(max_depth=3)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled = True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
