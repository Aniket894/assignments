Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import pandas as pd
import plotly.express as px

#plotly lib does not work properly

# Load tips dataset from the given URL
url = "https://github.com/mwaskom/seaborn-data/raw/master/tips.csv"
tips_dataset = pd.read_csv(url)

# Create a histogram using Plotly Express
histogram = px.histogram(
    tips_dataset,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Histogram of Total Bill Amount by Sex",
)

# Show the plot
histogram.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
import pandas as pd
import plotly.express as px

#plotly lib doest not work properly

# Load iris dataset from the given URL
url = "https://github.com/mwaskom/seaborn-data/raw/master/iris.csv"
iris_dataset = pd.read_csv(url)

# Create a scatter matrix plot using Plotly Express
scatter_matrix = px.scatter_matrix(
    iris_dataset,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset",
)

# Show the plot
scatter_matrix.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

ans - a distplot is used to display the distribution of a univariate set of observations. It combines a histogram with a kernel density estimate.

```python
import plotly.express as px
import numpy as np

# Generate some random data for demonstration
data = np.random.randn(1000)

# Create a distplot using Plotly Express
fig = px.histogram(data, nbins=50, marginal="rug", title="Distplot Example")

# Show the figure
fig.show()
```
