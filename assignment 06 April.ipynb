Q1. What is the mathematical formula for a linear SVM?

ans - The mathematical formula for a linear Support Vector Machine (SVM) can be defined as follows:

Given a set of labeled training data points (xᵢ, yᵢ), where xᵢ represents the input features and yᵢ is the corresponding class label (-1 or +1), the objective of a linear SVM is to find a hyperplane in the feature space that separates the data points of different classes with the maximum margin.

The decision function of a linear SVM can be written as:

f(x) = sign(w · x + b)

where:

f(x) represents the predicted class label for a new input vector x. w is the weight vector perpendicular to the hyperplane. · denotes the dot product between vectors. b is the bias term, which determines the offset of the hyperplane from the origin. The goal is to find the optimal values of w and b that maximize the margin while satisfying the following constraints:

yᵢ(w · xᵢ + b) ≥ 1

for all training data points (xᵢ, yᵢ), where ≥ represents the inequality constraint.

The margin is defined as the perpendicular distance between the hyperplane and the closest data points from either class. The optimal hyperplane is the one that maximizes this margin.

To find the optimal hyperplane, one can solve the following optimization problem:

minimize: 0.5 * ||w||² subject to: yᵢ(w · xᵢ + b) ≥ 1

where ||w|| represents the Euclidean norm of the weight vector w.

This problem can be transformed into a dual problem and solved using techniques such as quadratic programming to obtain the values of w and b that define the separating hyperplane.

Q2. What is the objective function of a linear SVM?

ans - ans - The objective function of a linear Support Vector Machine (SVM) is to find the optimal hyperplane that separates the data points of different classes with the maximum margin. In the case of a linear SVM, the objective function is typically formulated as a convex optimization problem.

The objective function of a linear SVM can be written as:

minimize: 0.5 * ||w||² + C * Σξᵢ

subject to: yᵢ(w · xᵢ + b) ≥ 1 - ξᵢ

where:

||w|| represents the Euclidean norm of the weight vector w. C is a regularization parameter that controls the trade-off between maximizing the margin and minimizing the training errors. A smaller value of C emphasizes a wider margin, while a larger value of C allows for more training errors. ξᵢ represents the slack variables, which allow for the possibility of misclassified points or points within the margin. The term Σξᵢ sums up the slack variables for all training data points. yᵢ is the class label of the training data point xᵢ. w · xᵢ + b is the decision function that determines the predicted class label for a given input vector xᵢ. The objective function consists of two terms. The first term, 0.5 * ||w||², encourages the maximization of the margin by minimizing the norm of the weight vector w. The second term, C * Σξᵢ, penalizes the training errors by adding a cost for each misclassified point or point within the margin.

By solving this optimization problem, the SVM finds the optimal values of w and b that define the hyperplane separating the classes while minimizing the training errors and maximizing the margin.

Q3. What is the kernel trick in SVM?

ans - ans - The kernel trick is a technique used in Support Vector Machines (SVMs) to handle non-linearly separable data by implicitly mapping the input features into a higher-dimensional feature space. It allows SVMs to efficiently solve non-linear classification problems without explicitly computing the transformation to the higher-dimensional space.

In a linear SVM, the decision boundary is a hyperplane that separates the data points of different classes. However, in many real-world scenarios, the data may not be linearly separable. The kernel trick addresses this limitation by introducing a kernel function that implicitly computes the dot product between the transformed feature vectors in the higher-dimensional space without explicitly transforming the data.

The kernel function, denoted as K(xᵢ, xⱼ), takes two input feature vectors xᵢ and xⱼ from the original input space and computes their similarity in the higher-dimensional feature space. The kernel function effectively calculates the dot product between the feature vectors, as if they were explicitly transformed into the higher-dimensional space.

By applying the kernel trick, the decision function of the SVM can be written as:

f(x) = sign(ΣαᵢyᵢK(xᵢ, x) + b)

where:

αᵢ is the Lagrange multiplier associated with each training data point (xᵢ, yᵢ). yᵢ is the class label of the training data point xᵢ. b is the bias term. K(xᵢ, x) is the kernel function that computes the similarity between xᵢ and x in the higher-dimensional feature space. Commonly used kernel functions include:

Linear Kernel: K(xᵢ, x) = xᵢ · x (the dot product of the input feature vectors). Polynomial Kernel: K(xᵢ, x) = (γxᵢ · x + c)ᵈ, where γ is a scaling factor, c is a constant, and d is the degree of the polynomial. Gaussian (RBF) Kernel: K(xᵢ, x) = exp(-γ||xᵢ - x||²), where γ is a scaling factor determining the kernel width. The advantage of the kernel trick is that it allows SVMs to effectively learn non-linear decision boundaries in the original input space without explicitly transforming the data. This avoids the computational burden and potential overfitting associated with explicit feature mapping while maintaining the benefits of SVM's convex optimization formulation.


Q4. What is the role of support vectors in SVM Explain with example

ans - In Support Vector Machines (SVMs), support vectors play a crucial role in defining the decision boundary and determining the classification of new data points. Support vectors are the data points that lie closest to the decision boundary (hyperplane) between different classes.

When training an SVM, the objective is to find the optimal hyperplane that maximizes the margin between the classes while minimizing the classification error. The support vectors are the data points that lie on or within the margin, as well as the misclassified points.

The key role of support vectors in SVM can be understood through the following example:

Imagine a two-class classification problem where the classes are linearly separable. In this case, the SVM will identify the support vectors as the data points that define the decision boundary or lie closest to it. These support vectors are crucial because they define the hyperplane and determine the separation of classes.

Let's say we have two classes, Class A and Class B, and a linear SVM is trained on the data. In the resulting decision boundary, some data points from Class A and Class B will lie on or within the margin. These data points will be identified as support vectors.

During the prediction phase, the classification of a new data point is determined by its similarity to the support vectors. The decision function of the SVM is evaluated for the new data point using the support vectors' information. The new data point is classified based on which side of the decision boundary it falls.

Support vectors also influence the robustness of the SVM model. As they are the data points closest to the decision boundary, they have the most influence on the model's generalization performance. Removing or altering support vectors can significantly affect the decision boundary and the model's accuracy.

Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

Hyperplane: A hyperplane in SVM is a decision boundary that separates the data points of different classes. In a binary classification problem, it is a line in 2D or a plane in 3D. In higher dimensions, it becomes a hyperplane. Example: Let's consider a 2D dataset with two classes, represented by blue circles and red crosses. The hyperplane is the line that separates the two classes.

Hyperplane

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Generate sample data
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
y = [-1] * 20 + [1] * 20

# Fit SVM model
clf = svm.SVC(kernel='linear')
clf.fit(X, y)

# Get the hyperplane parameters
w = clf.coef_[0]
b = clf.intercept_[0]
slope = -w[0] / w[1]  # Compute the slope of the hyperplane
xx = np.linspace(-5, 5)
yy = slope * xx - (b / w[1])  # Compute the equation of the hyperplane

# Plot the data points and the hyperplane
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.plot(xx, yy, 'k-')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hyperplane in SVM')
plt.axis('tight')
plt.show()
```

Marginal Plane: The marginal plane refers to the planes parallel to the hyperplane but at the closest distance from the data points of each class. These planes define the margin, which is the region between them. Example: In the same 2D dataset, the marginal planes are represented by the dashed lines on either side of the hyperplane. The region between these planes is the margin.

Marginal Plane

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Generate sample data
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
y = [-1] * 20 + [1] * 20

# Fit SVM model
clf = svm.SVC(kernel='linear')
clf.fit(X, y)

# Get the hyperplane parameters
w = clf.coef_[0]
b = clf.intercept_[0]
slope = -w[0] / w[1]  # Compute the slope of the hyperplane
xx = np.linspace(-5, 5)
yy = slope * xx - (b / w[1])  # Compute the equation of the hyperplane

# Get the support vectors and compute the margins
support_vectors = clf.support_vectors_
margin = 1 / np.linalg.norm(w)  # Compute the margin distance

yy_down = yy + slope * margin
yy_up = yy - slope * margin

# Plot the data points, hyperplane, and marginal planes
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.plot(xx, yy, 'k-', label='Hyperplane')
plt.plot(xx, yy_down, 'k--', label='Margin Plane')
plt.plot(xx, yy_up, 'k--', label='Margin Plane')
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hyperplane and Marginal Planes in SVM')
plt.legend()
plt.axis('tight')
plt.show()
```

Hard Margin: In a hard margin SVM, the goal is to find a hyperplane that perfectly separates the data points of different classes with no misclassifications. It requires the data to be linearly separable, meaning a hyperplane can completely separate the classes without any points falling into the margin. Example: Here, the dataset is linearly separable, and the hyperplane (solid line) perfectly separates the two classes without any misclassifications. The margin is maximized, and no data points lie within it.

Hard Margin

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Generate sample data
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
y = [-1] * 20 + [1] * 20

# Fit SVM model with a hard margin
clf = svm.SVC(kernel='linear', C=1e10)  # Setting C to a large value
clf.fit(X, y)

# Get the hyperplane parameters
w = clf.coef_[0]
b = clf.intercept_[0]
slope = -w[0] / w[1]  # Compute the slope of the hyperplane
xx = np.linspace(-5, 5)
yy = slope * xx - (b / w[1])  # Compute the equation of the hyperplane

# Plot the data points and the hyperplane
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.plot(xx, yy, 'k-')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hard Margin SVM')
plt.axis('tight')
plt.show()
```

Soft Margin: In contrast, a soft margin SVM allows for some misclassifications and data points to fall within the margin. It introduces a slack variable to relax the constraints and find a hyperplane that balances the margin size and misclassifications. Example: Consider a dataset that is not linearly separable. In a soft margin SVM, the hyperplane (solid line) allows for some misclassifications and data points within the margin. The margin is still maximized, but with some flexibility to accommodate outliers or overlapping data points.

Soft Margin

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Generate sample data
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
y = [-1] * 20 + [1] * 20

# Fit SVM model with a soft margin
clf = svm.SVC(kernel='linear', C=1.0)  # Setting C to a smaller value
clf.fit(X, y)

# Get the hyperplane parameters
w = clf.coef_[0]
b = clf.intercept_[0]
slope = -w[0] / w[1]  # Compute the slope of the hyperplane
xx = np.linspace(-5, 5)
yy = slope * xx - (b / w[1])  # Compute the equation of the hyperplane

# Get the support vectors and compute the margins
support_vectors = clf.support_vectors_
margin = 1 / np.linalg.norm(w)  # Compute the margin distance

yy_down = yy + slope * margin
yy_up = yy - slope * margin

# Plot the data points, hyperplane, and marginal planes
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.plot(xx, yy, 'k-', label='Hyperplane')
plt.plot(xx, yy_down, 'k--', label='Margin Plane')
plt.plot(xx, yy_up, 'k--', label='Margin Plane')
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Soft Margin SVM')
plt.legend()
plt.axis('tight')
plt.show()
```

```python
# Plot decision boundaries using two features
def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.23),
                         np.arange(y_min, y_max, 0.23))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.title('Decision Boundaries')
    plt.show()


    
```

```python
# Plot decision boundaries of the trained model
plot_decision_boundary(svc, X_train, y_train)
```
