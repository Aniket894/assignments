Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a Pandas DataFrame
flight_data = pd.read_csv('flight_price.xlsx - Sheet1.csv')


column_name = 'Price'

# Check if the specified column name exists in the DataFrame
if column_name not in flight_data.columns:
    print(f"Error: The column '{column_name}' does not exist in the dataset.")
else:
    # Plot a histogram of flight prices
    plt.figure(figsize=(10, 6))
    plt.hist(flight_data[column_name], bins=30, color='blue', edgecolor='black')
    plt.title('Distribution of Flight Prices')
    plt.xlabel('Flight Price')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
import pandas as pd


# Load the dataset into a Pandas DataFrame
flight_data = pd.read_csv('flight_price.xlsx - Sheet1.csv')


column_name = 'Price'

# Check if the specified column name exists in the DataFrame
if column_name not in flight_data.columns:
    print(f"Error: The column '{column_name}' does not exist in the dataset.")
else:
    # Find the minimum and maximum prices
    min_price = flight_data[column_name].min()
    max_price = flight_data[column_name].max()

    # Calculate the range
    price_range = max_price - min_price

    # Display the results
    print(f"The minimum price is: {min_price}")
    print(f"The maximum price is: {max_price}")
    print(f"The range of prices is: {price_range}")
```

Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
import pandas as pd


# Load the dataset into a Pandas DataFrame
playstore_data = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

app_column = 'App'
installs_column = 'Installs'

# Check if the specified column names exist in the DataFrame
if app_column not in playstore_data.columns or installs_column not in playstore_data.columns:
    print(f"Error: The columns '{app_column}' or '{installs_column}' do not exist in the dataset.")
else:
    # Clean the 'Installs' column and convert it to numeric
    playstore_data[installs_column] = playstore_data[installs_column].str.replace('[^\d.]', '', regex=True)
    playstore_data[installs_column] = pd.to_numeric(playstore_data[installs_column], errors='coerce')

    # Sort the DataFrame by installs in descending order
    sorted_data = playstore_data.sort_values(by=installs_column, ascending=False)

    # Select the top 10 apps
    top_10_apps = sorted_data.head(10)

    # Create a frequency table for the top 10 apps
    frequency_table = top_10_apps[[app_column, installs_column]]

    print("Top 10 Most Popular Apps:")
    print(frequency_table)
```
