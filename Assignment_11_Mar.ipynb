Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

ans  -  The t-test and z-test are both statistical tests used to make inferences about population parameters based on sample data, but they differ in their applications and assumptions.

1. T-test:

Use Case: The t-test is typically used when the sample size is small (usually less than 30) and the population standard deviation is unknown.
Example Scenario: Suppose you want to compare the average scores of two groups of students (Group A and Group B) on a math test. You have a sample of 20 students from each group, and you want to test if there is a significant difference in their mean scores.
2. Z-test:

Use Case: The z-test is appropriate when the sample size is large (typically n > 30) and/or the population standard deviation is known.
Example Scenario: Imagine you are conducting a study on the average height of a specific species of trees. You have a large sample size of 100 trees, and you know the population standard deviation of the tree heights. In this case, you might use a z-test to determine if a new treatment significantly affects the average height of the trees.

Q2: Differentiate between one-tailed and two-tailed tests.

ans -  1. One-Tailed Test:

Definition: A one-tailed test is a statistical test in which the critical region is located on only one side of the distribution of the test statistic.
Direction: It assesses whether the sample statistic is significantly greater than or significantly less than a population parameter, but not both.
Hypotheses Example: If you're investigating whether a new drug increases the average response time, your hypotheses might be:
Null Hypothesis (H0): The drug has no effect on response time (μ ≤ μ₀).
Alternative Hypothesis (H1): The drug increases response time (μ > μ₀), or the drug decreases response time (μ < μ₀).



2. Two-Tailed Test:

Definition: A two-tailed test is a statistical test in which the critical region is divided between both sides of the distribution of the test statistic.
Direction: It assesses whether the sample statistic is significantly different from a population parameter in either direction (greater than or less than).
Hypotheses Example: If you're examining whether a new teaching method affects test scores, your hypotheses might be:
Null Hypothesis (H0): The teaching method has no effect on test scores (μ = μ₀).
Alternative Hypothesis (H1): The teaching method has a significant effect on test scores (μ ≠ μ₀).


Key Differences:

Critical Region: In a one-tailed test, the critical region is on one side of the distribution, while in a two-tailed test, it is split between both sides.
Direction of Effect: One-tailed tests focus on whether the effect is in a specific direction, whereas two-tailed tests assess whether there is an effect in either direction.
Decision Making: For one-tailed tests, you reject the null hypothesis if the test statistic falls into the critical region on the specified side. For two-tailed tests, you reject the null hypothesis if the test statistic falls into either of the critical regions.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

ans - Type I and Type II Errors in Hypothesis Testing:

1. Type I Error:

Definition: Type I error occurs when the null hypothesis is incorrectly rejected when it is actually true. It represents a false positive or the probability of concluding that there is an effect when there is no actual effect.
Symbol: Denoted by α (alpha).
Example Scenario: Suppose you are conducting a medical test to determine if a patient has a rare disease. The null hypothesis (H0) is that the patient is healthy. A Type I error would occur if the test falsely indicates that the patient has the disease, leading to unnecessary treatments and stress for the patient.


2. Type II Error:

Definition: Type II error occurs when the null hypothesis is not rejected when it is actually false. It represents a false negative or the probability of failing to detect an effect that is genuinely present.
Symbol: Denoted by β (beta).
Example Scenario: Continuing with the medical test example, let's say the null hypothesis (H0) is that the patient is healthy, but in reality, the patient has the disease. A Type II error would occur if the test fails to detect the disease, leading to the patient not receiving necessary treatment, which could have serious consequences.
