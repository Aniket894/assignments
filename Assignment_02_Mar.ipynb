Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

ans - Matplotlib is a comprehensive library in Python used for creating static, interactive, and animated visualizations. It is widely employed for generating plots, charts, histograms, scatterplots, etc., aiding in data analysis and representation.

The library is used for various purposes, including:

Data Visualization: Matplotlib enables the creation of diverse visualizations to represent data, aiding in understanding patterns, trends, and relationships within datasets.

Customizable Plots: It offers extensive customization options, allowing users to modify almost every element of a plot, such as colors, labels, fonts, and more, to suit specific requirements.

Publication-Quality Figures: Matplotlib supports the creation of high-quality figures suitable for publications and presentations, ensuring professional-looking visualizations.

Integration with NumPy and Pandas: As Matplotlib integrates seamlessly with NumPy and Pandas, it simplifies the process of plotting data directly from these libraries.

Support for Various Plot Types: Matplotlib's Pyplot module provides functions to create a wide range of plots. Some of the commonly used plots include:

a. Line Plot: Shows data points connected by straight lines.

b. Scatter Plot: Displays points as individual markers, ideal for visualizing relationships between two variables.

c. Histogram: Represents the distribution of numerical data through bars, showcasing the frequency of values within certain ranges.

d. Bar Plot: Uses bars to represent categorical data, comparing values across different categories.

e. Pie Chart: Displays data as sectors of a circle, with each sector representing a proportion of the whole dataset.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

ans- A scatter plot is a type of data visualization that displays the relationship between two variables. Each point in the plot represents a pair of values, one for the x-axis and the other for the y-axis. It helps in identifying patterns, correlations, clusters, or trends within the data.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
plt.scatter(x, y, c='blue', alpha=0.7, edgecolors='black')  # 'c' sets the color, 'alpha' adjusts transparency, 'edgecolors' sets edge color

# Title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.grid(True)  # Optional: Add gridlines
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

ans - The subplot() function in Matplotlib is used to create multiple plots within a single figure. It helps in organizing and displaying different plots (or subplots) in a grid-like structure, allowing for a clear comparison between different datasets or visualizations.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 8))  # Optional: Adjust figure size
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, subplot 1
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, subplot 2
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, subplot 3
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, subplot 4
plt.plot(x4, y4)
plt.title('Line Plot 4')

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


ans - A bar plot is a graphical representation of categorical data with rectangular bars. It displays the comparisons among discrete categories. Bar plots are used to visualize and compare the values of different categories or groups. They help in illustrating trends, patterns, and disparities between different entities or groups in the data.

A horizontal bar plot, on the other hand, is similar to a vertical bar plot but with the bars displayed horizontally. This orientation is useful when dealing with long category names or when comparing categories with lengthy labels.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8, 6))

# Creating a vertical bar plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
plt.bar(company, profit, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')

# Creating a horizontal bar plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2
plt.barh(company, profit, color='salmon')
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

ans - A box plot, also known as a box-and-whisker plot, is used to display the distribution and range of a dataset. It shows the median, quartiles, and potential outliers in the data.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8, 6))

# Creating a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Values')

plt.show()
```
